# Simulated Annealing

```python
import numpy as np
import math as m
from math import *
```

```python
S = [3,2,1,5,10,36,33,12,11,6,18,50,4,7,17]
J = []
S1=S
T = []
T0=1000
fator = 0.5

a = np.linspace(-4.5,4.5,100)


for i in range(len(a)):
    Js = 20 + exp(-0.2*a[i]/100) - exp(cos(2*3.14*a[i])/100)
    J.append(Js)

f=[]
for i in range(len(a)):
    fx =20 + exp(-0.2*a[i]/100) - exp(cos(2*3.14*a[i])/100)
    f.append(fx)
print('conjunto S: \n',S)
print('\n')
print('função J: \n',J)
```

```python
def boltz():
    p = np.exp(-dJ/T0)
    return p
```

```python
import random
from random import choice

x0 = random.choice(S)
it = []
for n in range(20):
    for i in range(50):
        
        j = random.choice(S)
        
        indexj = S.index(j)
        indexX0 = S.index(x0)
        dJ = J[indexj]-J[indexX0]
        
        if(dJ==0):
            break        

        if(dJ<0):
            
            S.remove(x0)
            J.remove(J[indexX0])
            print('custo J=',J[indexj])
            
            
            x0=j
            
            
        elif(random.randrange(1,10)*0.1 < boltz()):
            
            x0=j    
            print('custo J=',J[indexj])
        
    T0 = T0*fator
    T.append(T0)
    it.append(n)
    print('Ti=',T0)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(it)), T, 'k')
plt.ylabel('T')
plt.xlabel('it')
plt.show()
```
